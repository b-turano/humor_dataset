```python
def no_burnham(title):
  if not re.search("Bo Burnham.*", title):
    return title
```

```python
def good_script(script_url):
  soup = bs(requests.get(script_url).text, "lxml")
  pars = soup.findAll("p")
  markers = ["(audience laughs)", "[audience laughs]", "(audience laughing)", "[audience laughing]", "[crowd laughs]", "[all laughing]", "[crowd laughing]", "[laughter]", "[laughter and applause]"]
  for p in pars:
    if p.get_text().lower() in markers:
      return script_url
    elif re.search("(laughs|laughing|laughter)", p.get_text().lower()):
      return script_url
    else:
      for marker in markers:
        if marker in p.get_text().lower():
          return script_url
      else:
          continue
```

```python
def get_script_id(link):
  discard = re.search(r"""https://scrapsfromtheloft.com/[0-9]{4}/[0-9]{2}/[0-9]{2}/""", link).span()
  try:
    dis2 = re.search(r"""-transcript/$""", link).span()
  except AttributeError:
    pass
  script_id = link[discard[1]:dis2[0]]
  return script_id
```

```python
def get_depth(i, j):
  """Get depth of sentence within script. i is the position of the sentence in the list (script); j is the 
  length of the list (script)"""
  return round((i/j)*100, 1)
```

```python
def process_text(sent):
  doc = nlp(sent)
  s = []
  NER = []
  for tok in doc:
    s.append(tok.text+"_"+tok.pos_)
  pos_tagged_s = " ".join(s)

  for ent in doc.ents:
    NER.append(ent.text+"_"+ent.label_)

  return pos_tagged_s, NER
```

```python
def get_label_and_depth(script):
  """This function saves the sentences of the script as either humourous or non-humourous, according to 
  whether a [LAUGHTER] tag appears in/follows them. It also computes the depth of the sentence in the script.
  Returns a list of tuples (sent, label, depth)."""
  lab_sents = []
  for i in range(len(script)):
    depth = get_depth(i+1, len(script))
    if script[i] == "[LAUGHTER]":
      continue
    try:
      if script[i+1] == "[LAUGHTER]":
        lab_sents.append((script[i], "Laugh", depth))
    except IndexError:
      break
    if "[LAUGHTER]" in script[i]:
      lab_sents.append((script[i], "Laugh", depth))
    else:
      lab_sents.append((script[i], "No-laugh", depth))

  return lab_sents

```
